## CSV ENCRYPTION / DECRYPTION of the file

```python
import cryptography.fernet as crypt
```

```python

# Fernet module is imported from the 
# cryptography package
import pandas as pd 
from cryptography.fernet import Fernet

  
# key is generated
key = Fernet.generate_key()



  
# value of key is assigned to a variable
f = Fernet(key)
print(f)

# the plaintext is converted to ciphertext
token = f.encrypt(b"welcome to geeksforgeeks")
token = f.encrypt(bytes("welcome to geeksforgeeks",'utf-8'))
  
# display the ciphertext
print(token)
  
# decrypting the ciphertext
d = f.decrypt(token)
  
# display the plaintext and the decode() method 
# converts it from byte to string
print(d.decode())
```
